```python
from TexSoup import TexSoup
with open('/content/gdrive/MyDrive/mCSE/resume4.tex') as f: #change path as required
  soup= TexSoup(f)
```

```python
section=[]
cont=[]
sec=list(soup.find_all('rSection')) #according to format 3, it is "section" tag
length=len(sec)
cmd=soup.count("itemize")
#print(cmd)
for i in range(length):
  p=str(sec[i].args[0])
  section.append( re.sub(r"[{}]+", "", p))
#print("sections: "+ str(section) )

cont=list(soup.find_all('rSection'))
```

```python
#bulleted resume or not?

item=[]
for i in range(len(cont)):
  item.append(cont[i].count("item"))
  #print(str(item[i]) + " bulleted items" + " in " + str(section[i]) + " section") 
dict1=dict(zip(section,item))
print("Sections:Bulleted Items")
pprint.pprint(dict1)
```

```python
#counting total links in resume
link_count=soup.count("href")
link=soup.find_all("href")
#print(str(link_count) + " links : " + str(link))

#taking only the url and not the pseudonames of the links
cleaned_link=[]
for i in range(link_count):
  clean_link=str(link[i].args[0])
  cleaned_link.append(re.sub(r"[{}]+","",clean_link))
#print(cleaned_link)

#checking professionalism and validity of the cleaned url list
url=str(cleaned_link)
if( re.findall("github.*",url) or re.findall("linkedin.*",url) or re.findall("behance.*",url) or  re.findall("gmail.com",url) or re.findall("iiitb.*",url) or  re.findall("outlook.*",url) ):
  isvalid=[]
  invalid=[]
  for i in range(len(cleaned_link)):
    valid=validators.url(cleaned_link[i])
    if valid==True:
      isvalid.append(cleaned_link[i])
    else:
      invalid.append(cleaned_link[i])

  print("Valid Professional Links: " + str(isvalid))
  #print("Invalid Professional Links: " + str(invalid))
else:
  print(url + " is unprofessional link")

```

```python
#checking validity of email addresses
email=[]
isv=[]
if(re.findall("gmail.com",url) or re.findall("iiitb.*",url) or  re.findall("outlook.*",url)):
  pattern = "^\S+@\S+\.\S+$"
  for j in range(len(invalid)):
    obj = re.search(pattern, str(invalid[j]))
    try:
      if obj.string == str(invalid[j]):
        email.append(str(invalid[j]))
        isv.append(validate_email(str(email[j]), verify=True))
        if(isv):
            print(str(email[j]) + " valid email")
    except:
        print(str(invalid) + " invalid email")
```
