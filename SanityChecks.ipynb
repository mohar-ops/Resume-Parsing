```python
!pip install texsoup
!pip install PyPDF2
import re 
import pprint
import PyPDF2
from PyPDF2 import PdfFileReader

#mount Google drive on colab
from google.colab import drive

drive.mount('/content/gdrive')
```

```python
#open resume in PDF version for number of pages
pdfObj = open('/content/gdrive/MyDrive/mCSE/resume4.pdf','rb')
pdfRead = PyPDF2.PdfFileReader(pdfObj)
# Get total number of pages in the resume
pages = pdfRead.numPages
print("Pages in Resume: "+ str(pages))
```

```python
#form soup tree from tex file
from TexSoup import TexSoup
with open('/content/gdrive/MyDrive/mCSE/resume4.tex') as f: #path of LaTex file on google drive
  soup= TexSoup(f)
#print(soup)
```

```python
#cleaning and extracting sections
import re
section=[]

sec=list(soup.find_all('rSection'))
length=len(sec)

for i in range(length):
  p=str(sec[i].args[0])
  section.append( re.sub(r"[{}]+", "", p))
  
print("Sections in resume: "+ str(section))

#type(section)->list
```
